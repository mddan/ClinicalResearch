```python
# Read SAS file in chunks
sasfile1 = "../raw_data/test_data/app_0618_rs_exc_g2ch.sas7bdat"
chunk_size=5000
data_chunked = pd.read_sas(sasfile1, chunksize=chunk_size, iterator=True, format = 'sas7bdat', encoding="latin-1")

# Holds data chunks 
df_list = [] 

# Extracting and appending chunks
for chunk_count, chunk in enumerate(data_chunked):
    if chunk_count == 0:
        clear_output(wait=True)
    df_list.append(chunk)
    print(f"Update: Extracted chunk #{chunk_count+1} ({len(chunk)} rows)")
df = pd.concat(df_list)
print(f"Finished concatenating all {chunk_count + 1} chunks!")
```
